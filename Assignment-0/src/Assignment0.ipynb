```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/drive/My Drive/COMPUTER VISION/img3.jpg')
#cv2_imshow(img)
(row, col, channels) = img.shape
print(row, col, channels)
gray = np.zeros((row,col))
b, g, r = cv2.split(img)
#img1_grey = (0.299*img[...,2] +  0.587*img[...,1] + 0.114*img[...,0])
'''
OR CAN SPLIT IT THIS WAY

b = img[...,0]
g = img[...,1]
r = img[...,2]
'''
gray = (r*0.2989 + g*0.587 + 0.114*b)
cv2_imshow(gray)

filename = '/content/drive/My Drive/COMPUTER VISION/gray.jpg'
cv2.imwrite(filename, gray) 
```
